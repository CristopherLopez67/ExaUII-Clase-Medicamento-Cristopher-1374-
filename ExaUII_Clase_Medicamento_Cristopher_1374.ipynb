```python
class Medicamento:
  def __init__(self, nombre, compuesto, dosis):
    self.nombre = nombre
    self.compuesto = compuesto
    self.dosis = dosis

  def leerdatos(self):
    print("Por favor, ingrese los datos del medicamento:")
    self.nombre = input("Ingrese el nombre del medicamento: ")
    self.compuesto = input("Ingrese el compuesto del medicamento: ")
    self.dosis = input("Ingrese la dosis del medicamento: ")

  def mostrardatos(self):
    print("Datos del Medicamento:")
    print(f"Nombre: {self.nombre}")
    print(f"Compuesto: {self.compuesto}")
    print(f"Dosis: {self.dosis}")

# Fin de la clase Medicamento

# Creando los objetos med1, med2

# Inicializamos los objetos
med1 = Medicamento("", "", "")
med2 = Medicamento("", "", "")

print("Medicamento Numero 1")

med1.leerdatos()
med1.mostrardatos()
print("-------------------------------------------------")

print("Medicamento Numero 2")
med2.leerdatos()
med2.mostrardatos()
print("-------------------------------------------------")

print("programa realizado por Cristopher Lopez NL = 33")
```
